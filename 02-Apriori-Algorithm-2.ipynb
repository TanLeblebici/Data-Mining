# 3. Convert table to pivot table with index column = 'Invoice & Items(Columns) = Desc and number of table

```python
df_pivot = pd.pivot_table(data = df,
                               index = 'InvoiceNo',
                               columns = 'Description',
                               values = 'Quantity',
                               aggfunc = 'sum',
                               fill_value = 0)
df_pivot
```

```python
df_pivot.max(numeric_only = 'true')
```

```python
df_pivot.drop(columns = ['POSTAGE'],inplace=True)
df_pivot.drop(columns = ['wrongly coded 20713'],inplace=True)
df_pivot.drop(columns = ['wrongly coded 23343'],inplace=True)
df_pivot.drop(columns = ['wrongly marked 23343'],inplace=True)
df_pivot.drop(columns = ['wrongly marked'],inplace=True)
df_pivot.drop(columns = ['wrongly sold (22719) barcode'],inplace=True)

df_pivot
```

```python
def dataframe_to_list_of_lists(df):
    transactions = []
    for index, row in df.iterrows():
        items = [column for column in df.columns if row[column] > 0]
        transactions.append(items)
    return transactions
```

```python
transactions = dataframe_to_list_of_lists(df_pivot)
```

```python
for i in range(10):
    print(str(i) + "-" + str(transactions[i]))
```

# 4. import apyori and run apriori algorithm

```python
rules = apriori(transactions, 
                min_support = 0.01, 
                min_confidence = 0.5, 
                min_lift = 3, 
                min_length = 2, 
                max_length = 3)
```

```python
results = list(rules)
```

```python
results
```

```python
def convert_dataFrame(results):
    lhs = [', '.join(result[2][0][0]) for result in results]  # Join multiple LHS items by comma
    rhs = [', '.join(result[2][0][1]) for result in results]  # Join multiple RHS items by comma
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
```

```python
resultFrame = pd.DataFrame(convert_dataFrame(results),columns = ["Left Hand Side","Right Hand Side","Support","Confidence","Lift"])
resultFrame
```
